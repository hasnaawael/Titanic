## Age Feature

#### guess age for people with no age based on Pclass , Gender
    

```python
### creat empty array for guessed age 
guess_ages = np.zeros((2,3))
guess_ages
```

```python
###loop over sex and pclass to calculate age
for dataset in combine:
    ###Get in trian data and ignore null values
    for i in range(0,2):### 0, 1 === Gender
        for j in range(0,3): ### 0, 1, 2 === pclass
            guess_df = dataset[(dataset['Sex'] == i)&  
                              (dataset['Pclass']==j+1)]['Age'].dropna()
            
            age_guess = guess_df.median()### get this values
            guess_ages[i,j] = int (age_guess/0.5+0.5)*0.5###convert float to nearest .5 age
            
    ### loop in test data to fill null values       
    for i in range(0,2):
        for j in range(0,3):
            dataset.loc[ (dataset.Age.isnull())&(dataset.Sex ==i)&(dataset.Pclass==j+1), 'Age']=guess_ages[i,j]
    dataset['Age'] = dataset['Age'].astype(int)
```

```python
train_df.sample(5)
```

```python
### divid age to sectors 
###هقسم السن من كبير للصغير 
train_df['AgeBand'] = pd.cut(train_df['Age'] , 5) ### 5 sections
train_df[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=False)
```

```python
### replace age ordainals based on these bands
for dataset in combine:
    dataset.loc[ dataset['Age'] <= 16 , 'Age'] = 0 ## all age under 16 and 16 will be 0
    dataset.loc[ (dataset['Age'] > 16 ) & (dataset['Age'] <= 32 ) , 'Age'] = 1
    dataset.loc[ (dataset['Age'] > 32 ) & (dataset['Age'] <= 48 ) , 'Age'] = 2
    dataset.loc[ (dataset['Age'] > 48 ) & (dataset['Age'] <= 64 ) , 'Age'] = 3
    dataset.loc[ (dataset['Age'] > 64) ,'Age'] = 4
```

```python
train_df.sample(5)
```

## Embarked Feature|

#### fill null values with the most common Embarked

```python
### to know the most common Embarked
freq_port = train_df.Embarked.dropna().mode()[0] ### without null value
freq_port
```

```python
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].fillna(freq_port)### fill it with S
train_df[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

#### convert it to numeric

```python
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].map({'S': 0, 'C': 1, 'Q':2}).astype(int)
```

## Fare Feature
#### in test data fill null with median

```python
test_df['Fare'].fillna(test_df['Fare'].dropna().median(), inplace=True)
```

```python
test_df.sample(5)
```

#### divide it into 4 bands

```python
train_df['FareBand'] = pd.qcut(train_df['Fare'], 4)
train_df[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean().sort_values(by='FareBand', ascending=False)
```

```python
### replace fare ordainals based on these bands
for dataset in combine:
    dataset.loc[ dataset['Fare'] <= 7.91 , 'Fare'] = 0 ## all age under 7.91 and 7.91 will be 0
    dataset.loc[ (dataset['Fare'] > 7.91 ) & (dataset['Fare'] <= 14.454 ) , 'Fare'] = 1
    dataset.loc[ (dataset['Fare'] > 14.454 ) & (dataset['Fare'] <= 31 ) , 'Fare'] = 2
    dataset.loc[ (dataset['Fare'] > 31 ,'Fare')] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)
```

# Now build model

#### split it to train and test

```python
X_train = train_df.drop("Survived", axis=1)
Y_train = train_df["Survived"]
X_test  = test_df.drop("PassengerId", axis=1).copy()
```

### Logistic Regression

#### fit model

```python
from sklearn.linear_model import LogisticRegression
log= LogisticRegression()
log.fit(X_train, Y_train)
```

```python
###predict
Y_pred = log.predict(X_test)
acc_log = round(log.score(X_train, Y_train) * 100, 2)
acc_log
```
